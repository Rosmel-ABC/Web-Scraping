## 1: Extracci√≥n de valores (Nombre de las laptops)

```python
# Accedemos a la ruta del  DRIVER previamente descargados y descomprimido
driver = webdriver.Chrome("C:/Users/Usuario/Downloads/chromedriver_win32/chromedriver.exe")

#le pasamos la url de la pagina a scrapear
driver.get("https://www.falabella.com.pe/falabella-pe/category/cat40712/Laptops") 

#Creamos una lista vac√≠a donde se almacenar√° los datos obtenidos
nombres_productos = []



# Extraccion de nombres
    ## Iteramos para poder extraer mas datos, en este caso los 9 primeros productos
    
for i in range(1,10):  
    
    #Esta direccion lo obtenemos de ponernos sobre el nombre del producto e inspeccionamos el elemento
    # y copiamos el  X_path completo
    
    # Hacemos que itere para poder acceder a mas datos,  usando format(i) en {}
    x_path = "//html/body/div[1]/div/div[2]/div[2]/section[2]/div/div[3]/div[{}]/div/div[2]/a/span[1]".format(i)
    
    #Los datos obtenidos lo agregamos  la lista "nombres_productos" previamente creada
    nombres_productos.append(driver.find_elements(By.XPATH, value = x_path)[0].text)   #esto es una lista 
```

```python
#Mostramos los datos de la lista
nombres_productos
```

```python
#vemos la cantidad de datos de la lista
len(nombres_productos)
```

## 2: Extracci√≥n de valores de los precios con descuento de las laptops

```python
# Accedemos a la ruta del  DRIVER previamente descargados y descomprimido
driver = webdriver.Chrome("C:/Users/Usuario/Downloads/chromedriver_win32/chromedriver.exe")

#le pasamos la url de la pagina a scrapear
driver.get("https://www.falabella.com.pe/falabella-pe/category/cat40712/Laptops") 
            
#Creamos una lista vac√≠a donde se almacenar√° los datos obtenidos
precios_productos_descuento = []


# Extraccion de nombres
    ## Iteramos para poder extraer mas datos, en este caso los 9 primeros productos
for i in range(1,10):  

    try:
    #Esta direccion lo obtenemos de ponernos sobre el precio con descuento del producto y inspeccionamos el elemento
    # y copiamos el  X_path completo
    
        # Hacemos que itere para poder acceder a mas datos,  usando format(i) en {}
        x_path = "//html/body/div[1]/div/div[2]/div[2]/section[2]/div/div[3]/div[{}]/div/a/div[1]/ol/li[1]/div/span".format(i)           
       
        #Los datos obtenidos lo agregamos  la lista "precios_productos_descuento" previamente creada
        precios_productos_descuento.append(driver.find_elements(By.XPATH, value = x_path)[0].text)   #esto es una lista 

    except:
        
        #En caso no existe este dato lo completa con "-"
        precios_productos_descuento.append("-") 
```

## <center><h1><b>Extracci√≥n de datos todo en uno</b></h1></center>

```python
# Accedemos a la ruta del  DRIVER previamente descargados y descomprimido
driver = webdriver.Chrome("C:/Users/Usuario/Downloads/chromedriver_win32/chromedriver.exe")

#le pasamos la url de la pagina a scrapear
driver.get("https://www.falabella.com.pe/falabella-pe/category/cat40712/Laptops") 

#Creamos listas vac√≠a donde se almacenar√°n los datos obtenidos
nombres_productos = []
precios_productos = []
precios_productos_descuento = []
precios_3 = []

#creamos un contador cuantas veces debe pasar de p√°gina
contador = 0


while contador <2:
    for i in range(1,30):  
            #1
            x_path = "//html/body/div[1]/div/div[2]/div[2]/section[2]/div/div[3]/div[{}]/div/div[2]/a/span[1]".format(i)
            nombres_productos.append(driver.find_elements(By.XPATH, value = x_path)[0].text)   #esto es una lista 

            
            #2----------------------------------------------------------
            try:
                x_path = "//html/body/div[1]/div/div[2]/div[2]/section[2]/div/div[3]/div[{}]/div/a/div[1]/ol/li[1]/div/span".format(i)           
                precios_productos_descuento.append(driver.find_elements(By.XPATH, value = x_path)[0].text)   #esto es una lista 

            except:
                precios_productos_descuento.append("-") 
                
                
            #3 --------------------------------------------------------------
            try :
                x_path = "//html/body/div[1]/div/div[2]/div[2]/section[2]/div/div[3]/div[{}]/div/a/div[1]/ol/li[2]/div/span".format(i)                  
                precios_productos.append(driver.find_elements(By.XPATH, value = x_path)[0].text)   #esto es una lista 
            
            except:
                precios_productos.append("-")


            #4 ----------------------------------------------------------
            try : 
                x_path = "//html/body/div[1]/div/div[2]/div[2]/section[2]/div/div[3]/div[{}]/div/a/div[1]/ol/li[3]/div/span".format(i)        
                precios_3.append(driver.find_elements(By.XPATH, value = x_path)[0].text)   #esto es una lista 
            except:
                precios_3.append("-")
                
                
    #5 codigo para pasar de p√°gina            
    x_path = "//html/body/div[1]/div/div[2]/div[2]/section[2]/div/div[1]/div/div[2]/div/div/button"
               
    driver.find_elements(By.XPATH, value = x_path)[0].click()
    contador = contador + 1
```

## CREAMOS DATAFRAME CON LAS LISTAS CREADAS DE PRODUCTO Y PRECIO

```python
import numpy as np

data = pd.DataFrame(nombres_productos , columns = ["Productos"])
data["Precio con descuento"] = precios_productos_descuento
data["Precio"] = precios_productos
data["Precio Anterior"] = precios_3
```
